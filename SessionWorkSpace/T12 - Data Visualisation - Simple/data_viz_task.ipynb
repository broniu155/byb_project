# Data visualisation task 

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
cars_df = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# create the dataframe for requested manufactures: Audi,Hyundai,Suzuki,Toyota
four_car_manu = cars_df[cars_df['Manufacturer'].isin(['Audi', 'Hyundai', 'Suzuki', 'Toyota'])]
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


```python
# Boxplot for revs per mile for required car manufacturers
sns.boxplot(data=four_car_manu, x='Manufacturer', y='Rev.per.mile',whis=(0, 100))#whis added to cover all of the outliers
```

**Boxplot answer:**


```python
print('Manufacturer with the highest revs per mile is Toyota')
```

### 2. Histogram

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

```python
# Histogram of MPG
sns.histplot(cars_df['MPG.city'], bins=20, color='blue',kde=True, label='MPG City')# More bins and color added for better visibility
sns.histplot(cars_df['MPG.highway'], bins=20, color='green', kde=True, label='MPG Highway')# More bins and color added for better visibility
plt.title('Comparison of MPG in City vs Highway')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
plt.close()
```

**Histogram answer:**

```python
print('More fuel efficient is to drive on the highway')
```

### 3. Lineplot

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(data=cars_df, x='Wheelbase', y='Turn.circle')
plt.ylabel('Turning Circle')
plt.show()
plt.close()
```

**Lineplot answer:**  


```python
print('The higher Wheelbase there is higher Turning Circle')
```

### 4. Barplot

**Barplot question:**  
Does a larger car mean more horsepower?



```python
# Group data by'Type' and mean (average) for each type across all columns
group_by_type = round(cars_df.groupby('Type')['Horsepower'].mean(),2)
group_by_type.sample(5)

#Sort the values and reset index from original dataframe
sorted_by_horspower = group_by_type.sort_values(ascending=True).reset_index()

sorted_by_horspower.sample(5)
```

```python
# Barplot for mean horsepower
sns.barplot(data=sorted_by_horspower, x='Type', y='Horsepower')
plt.title('Mean Horsepower by Car Type')
plt.xlabel('Type')
plt.ylabel('Horsepower')
plt.show()
plt.close()
```

**Barplot answer:**

```python
print('On average, larger cars tend to have more horsepower')
```
